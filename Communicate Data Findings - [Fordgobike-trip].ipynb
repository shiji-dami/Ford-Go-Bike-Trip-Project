### DATA CLEANING

```python
# creating a copy of the dataset
ford_bike_clean = ford_bike.copy()
```

```python
# removing null rows from the dataset
ford_bike_clean.dropna(axis = 0 , inplace = True)
```

```python
# Extracting the 'member_age' of members from the 'member_birth_year' colunm
ford_bike_clean['member_age'] = 2022 - ford_bike_clean['member_birth_year']
```

```python
# changing the datatype of the start_station_id, end_station_id, 'member_age' to int
for b in ['start_station_id', 'end_station_id','member_age']:
    ford_bike_clean[b] = ford_bike_clean[b].astype(int)
```

```python
# changing the datatype of the member_gender, end_station_id to int
for c in ['member_gender', 'user_type']:
    ford_bike_clean[c] = ford_bike_clean[c].astype('category')
```

```python
# changing the datatype of the `start_time`, `end_time` to datetime
ford_bike_clean[['start_time','end_time']] = ford_bike_clean[['start_time','end_time']].apply(pd.to_datetime, errors='coerce')
```

```python
# adding a new column 'duration_min' 
ford_bike_clean['duration_min'] = ford_bike_clean['duration_sec']/60
```

```python
# adding a new column 'duration_hour' 
ford_bike_clean['duration_hour'] = ford_bike_clean['duration_sec']/3600
```

```python
# to extract month of the 'start_week' column
ford_bike_clean['start_month'] = pd.to_datetime(ford_bike_clean['start_time']).dt.month_name()
```

```python
ford_bike_clean['start_hour'] = pd.to_datetime(ford_bike_clean['start_time']).dt.hour
```

```python
# to extract day of the 'start_week' column
ford_bike_clean['start_week'] = pd.to_datetime(ford_bike_clean['start_time']).dt.day_name()
```

```python
# dropping unwanted columns such as `start_station_latitude`, `start_station_longitude`, `end_station_latitude`, `end_station_longitude`. 
ford_bike_clean.drop(['start_station_latitude', 'start_station_longitude', 'end_station_latitude', 'end_station_longitude', 'member_birth_year'], axis = 1, inplace = True)
```

#### What is the gender distribution?

```python
# Recalculating the type_counts just to have clarity.
n_user = ford_bike_clean['member_gender'].value_counts().sum()
user_type_counts = ford_bike_clean['member_gender'].value_counts()

# plotting the distribution of member gender
base_color = sns.color_palette()[0]
order = user_type_counts.index
sns.countplot(data=ford_bike_clean, x='member_gender', color=base_color, order=order);
 

# get the current tick locations and labels
locs, labels = plt.xticks() 

# loop through each pair of locations and labels
for loc, label in zip(locs, labels):

    # get the text property for the label to get the correct count
    count = user_type_counts[label.get_text()]
    pct_string = '{:0.1f}%'.format(100*count/n_user)

    # print the annotation just below the top of the bar
    plt.text(loc, count+2, pct_string, ha = 'center', color = 'black')
    
# Labeling the axes
plt.xlabel('member_gender')
plt.title('Distribution of member gender')

# Dsiplay the plot
plt.show()
```
